````
import pandas as pd
import os

# ดึงข้อมูลจากทุกไฟล์ CSV ที่เกี่ยวกับ stop-and-search และเพิ่มชื่อเมืองเข้าไปในแต่ละตาราง
def load_all_data(folder_path):
    combined_data = []
    month_dirs = os.listdir(folder_path)

    for month in month_dirs:
        full_month_path = os.path.join(folder_path, month)
        filenames = os.listdir(full_month_path)

        for name in filenames:
            # เลือกเฉพาะไฟล์ที่เป็น .csv และมีคำว่า 'stop-and-search' ในชื่อ
            if name.endswith(".csv") and 'stop-and-search' in name:
                file_path = os.path.join(full_month_path, name)
                df = pd.read_csv(file_path)

                # ตัดชื่อเมืองออกมาจากชื่อไฟล์
                city_name = name.replace('-stop-and-search.csv', '').split('-', 2)[-1]
                df['Police Force'] = city_name  # เพิ่มชื่อเมืองลงในคอลัมน์ใหม่
                combined_data.append(df)

    return combined_data

# นับจำนวนการถูกตรวจค้นของบุคคลที่ระบุว่าเป็นเชื้อสายเอเชีย
def count_ethnicity_asian(data_list):
    summary = {}

    for table in data_list:
        for _, row in table.iterrows():
            ethnicity = str(row.get('Self-defined ethnicity', ''))
            if 'Asian' in ethnicity:
                city = row['Police Force']
                summary[city] = summary.get(city, 0) + 1

    return summary

# -------- จุดเริ่มต้นของโปรแกรม --------
root_path = '/content/drive/My Drive/Python&Dataviz2025/Data/'

# โหลดข้อมูลทุกเดือน
all_data = load_all_data(root_path)

# ประมวลผลจำนวนคนเชื้อสายเอเชียที่ถูกตรวจค้น
asian_summary = count_ethnicity_asian(all_data)

# หาค่ามากสุดและน้อยสุด
top_city = max(asian_summary, key=asian_summary.get)
bottom_city = min(asian_summary, key=asian_summary.get)

# แสดงผลลัพธ์
print(">>> จำนวนการตรวจค้นบุคคลเชื้อสายเอเชียในแต่ละเมือง (รวมหลายเดือน):\n")
for city in sorted(asian_summary, key=asian_summary.get, reverse=True):
    print(f"{city}: {asian_summary[city]}")

print("\n=== สรุปข้อมูล ===")
print(f"เมืองที่มีการเรียกค้นคนเชื้อสายเอเชียมากที่สุด: {top_city} ({asian_summary[top_city]})")
print(f"เมืองที่มีการเรียกค้นคนเชื้อสายเอเชียน้อยที่สุด: {bottom_city} ({asian_summary[bottom_city]})")
````

HW 5 (เดี่ยว) ***แก้จาก code เดิม*** ยอดรวมทุกเมือง เดือนไหนมีจำนวนผู้ถูกเรียกค้นที่เป็นคน Asia น้อยที่สุด และ มากที่สุด

````
import pandas as pd
import os

# โหลดข้อมูลจากทุกเมืองและทุกเดือน
def load_all_data(folder_path):
    combined_data = []
    month_dirs = os.listdir(folder_path)

    for month in month_dirs:
        full_month_path = os.path.join(folder_path, month)
        filenames = os.listdir(full_month_path)

        for name in filenames:
            if name.endswith(".csv") and 'stop-and-search' in name:
                file_path = os.path.join(full_month_path, name)
                df = pd.read_csv(file_path)

                # เพิ่มชื่อเดือนจากชื่อโฟลเดอร์เข้าไป
                df['Month'] = month
                combined_data.append(df)

    return pd.concat(combined_data, ignore_index=True)

# -------- เริ่มต้นโปรแกรม --------
root_path = '/content/drive/My Drive/Python&Dataviz2025/Data/'

# โหลดข้อมูลทุกเดือน
all_data = load_all_data(root_path)

# กรองเฉพาะคนเชื้อชาติ Asian
asian_data = all_data[all_data['Self-defined ethnicity'].str.contains('Asian', na=False)]

# นับจำนวนคน Asian แยกตามเดือน
monthly_counts = asian_data.groupby('Month').size().reset_index(name='Asian_Count')

# หาค่าน้อยสุดและมากสุด
min_row = monthly_counts.loc[monthly_counts['Asian_Count'].idxmin()]
max_row = monthly_counts.loc[monthly_counts['Asian_Count'].idxmax()]

# แสดงผลลัพธ์
print(">>> จำนวนการเรียกค้นบุคคลเชื้อชาติ Asian แยกตามเดือน:\n")
print(monthly_counts.sort_values(by='Asian_Count', ascending=False))

print("\n=== สรุปข้อมูล ===")
print(f"เดือนที่มีคนเอเชียน้อยที่สุด: {min_row['Month']} ({min_row['Asian_Count']} คน)")
print(f"เดือนที่มีคนเอเชียมากที่สุด: {max_row['Month']} ({max_row['Asian_Count']} คน)")
````
